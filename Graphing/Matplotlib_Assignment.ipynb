```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
p_squares = list(np.arange(1,11) ** 2)
```

```python
# Default
plt.plot(p_squares, color = "red")
plt.ylabel("Squares")
plt.xticks(np.arange(len(p_squares)), np.arange(1, len(p_squares)+1))
plt.show()
```

```python
# Scatterplot + line combined
plt.scatter(list(np.arange(0,10)), p_squares, color = "red")
plt.plot(p_squares, color = "green")
plt.ylabel("Squares")
plt.xticks(np.arange(len(p_squares)), np.arange(1, len(p_squares)+1))
plt.show()
```
